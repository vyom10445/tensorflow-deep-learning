```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create features (using tensors)
x = tf.constant([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

# Create labels (using tensors)
y = tf.constant([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

# Visualize it
plt.scatter(x, y);
```

```python
# take a single example of X
input_shape = x[0].shape

# take a single example of y
output_shape = y[0].shape

input_shape, output_shape # these are both scalars (no shape)
```

```python
# looking at the single examples invidually
x[0], y[0]
```

```python
#create a model using Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1) #only one unit and one layer cos for one input were expecting one output
])

#compile the model
model.compile(loss=tf.keras.losses.mae,#mean average error
              optimizer=tf.keras.optimizers.SGD(),#stochastic gradient descent
              metrics=["mae"]
              )

#fit the model
model.fit(tf.expand_dims(x, axis=-1), y, epochs=100)
```

```python
#better results than 5 epochs . we're trying to improve the model step by step. step 1 was above, trying to increase epochs
```

```python
x,y
```

```python
model.predict(tf.constant([17.0]))
```

```python
#mustve been 27 (y=x+10)
```

```python
#now adding more hidden layers to model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(1)
])

model.compile(loss = tf.keras.losses.mae ,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
model.fit(tf.expand_dims(x, axis=-1), y, epochs=100)
```

```python
model.predict(tf.constant([17.0]))
```

```python
#not accurate still , worse result than before idk
#the model is overfitted
```

```python
#maybe reducing no. of layers and units , and tweaking activation function , optimizer , lr , lets work and find our best model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation=None),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),#adam default lr was 0.0001 but we took 0.1 cos model was worse with 0.0001,
              metrics=["mae"])

model.fit(tf.expand_dims(x, axis=-1), y, epochs=100)
```

```python
model.predict(tf.constant([17.0]))
```
